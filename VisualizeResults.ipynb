```python
import pandas as pd
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('results.csv',index_col=0)
df = df.fillna('SQLNet').replace(['ca', 'baseline', 'train_emb'],['Column Attention','Baseline','Train Embeddings'])
```

```python
def plot(y, save=False):
    sns.lineplot(x='Epoch',y=y, hue='Type',data=df)
    plt.title(y)
    plt.xlabel('Epoch')
    plt.ylabel(y)
    if save: plt.savefig('Figures/'+y+'.png')
```

```python
plot('Loss',save=True)
```

```python
plot('Train Acc',save=True)
```

```python
plot('Val Acc',save=True)
```
