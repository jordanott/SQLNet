```python
import pandas as pd
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('results.csv',index_col=0)
df.head()
```

```python
def plot(y, save=False):
    sns.lineplot(x='Epoch',y=y, hue='Type',data=df)
    plt.title('Training '+y)
    plt.xlabel('Epoch')
    plt.ylabel(y)
    if save: plt.savefig('Figures/'+y+'.png')
```

```python
plot('Loss')
```

```python
plot('Train Acc')
```

```python
plot('Val Acc')
```
